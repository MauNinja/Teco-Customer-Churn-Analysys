```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by SeniorCitizen")
plt.show()
```

```python
# Calculate the counts for each SeniorCitizen and Churn category
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
# Convert the counts to percentages
counts_percent = counts.apply(lambda x: (x / x.sum()) * 100, axis=1)

# Plot the stacked bar chart
ax = counts_percent.plot(kind='bar', stacked=True, figsize=(4,4), color=['#1f77b4', '#ff7f0e'])

# Add percentages on top of each bar segment
for i in ax.containers:
    ax.bar_label(i, label_type='center', fmt='%.1f%%')

# Add title and labels
plt.title('Churn by Senior Citizen (Percentage)')
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen')

# Show the legend
plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9))

# Display the plot
plt.show()
```

```python
# List of columns for which we want to create countplots
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 
           'StreamingTV', 'StreamingMovies']

# Number of columns to show in the subplot grid
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols 
# Create a figure and subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))

# Flatten axes array for easier iteration
axes = axes.flatten()

# Loop through the columns and create countplots
for i, col in enumerate(columns):
    sns.countplot(x = col, data = df, ax=axes[i], hue = df["Churn"] )
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove any unused subplots
for j in range(i + 1,len(axes)):
    fig.delaxes(axes[j])

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

#### In general, customers who do not have additional services like OnlineSecurity, TechSupport, or DeviceProtection tend to churn at higher rates. Additionally, fiber-optic internet service shows a higher churn rate compared to DSL. The data suggest a pattern where having fewer bundled services correlates with increased churn.
